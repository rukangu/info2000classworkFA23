```python
import requests
import pandas as pd
```

```python
base_url = 'https://www.alphavantage.co/query'
my_symbol = 'MU'
my_key = open('api_key').read()
my_params= {'function':'TIME_SERIES_DAILY', 'symbol':my_symbol, 'apikey':my_key } # creating a dictionary of parameters
```

```python
# send to the server

response = requests.get(base_url, params=my_params)
```

```python
response.headers
```

```python
response.text
```

```python
# convet to python dict

response.json()['Time Series (Daily)']
```

```python
df = pd.DataFrame.from_records(response.json()['Time Series (Daily)']).T
df
```
