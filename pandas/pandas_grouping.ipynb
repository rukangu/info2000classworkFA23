```python
import pandas as pd
import numpy as np
```

```python
student_performance_df = pd.read_csv('StudentsPerformance_pandas.csv', header=2)
student_performance_df.head()
display(student_performance_df.info())
display(student_performance_df.describe())
```

```python
# clean our data

new_columns = []
for col in student_performance_df.columns:
    new_columns.append(col.replace("/","_").replace(" ", "_"))
new_columns
```

```python
student_performance_df.columns = new_columns
```

```python
student_performance_df.head()
```

```python
student_performance_df.groupby("gender").sum(numeric_only=True)
```

```python
student_performance_df['average'] = (student_performance_df.math_score + student_performance_df.writing_score + student_performance_df.reading_score)/3

def grade(row):
    average = row.average
    if average >= 90:
        return 'A'
    elif average >= 80 and average <90:
        return 'B'
    else:
        return 'Pass'


student_performance_df['grades'] = student_performance_df.apply(grade, axis=1)
student_performance_df.head()
```

```python
student_performance_df.groupby(['gender','grades']).mean(numeric_only = True)
```

```python
female_performance = student_performance_df.groupby(['gender','grades']).mean(numeric_only = True).loc['female']
female_performance
```

```python
male_performance = student_performance_df.groupby(['gender','grades']).mean(numeric_only = True).loc['male']
male_performance
```

```python
student_performance_df.groupby(['parental_level_of_education', 'lunch']).mean(numeric_only = True).sort_values('average')
```
